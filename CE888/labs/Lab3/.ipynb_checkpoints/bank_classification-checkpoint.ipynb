```python
%matplotlib inline
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.metrics import f1_score
import itertools
import sklearn
from sklearn.model_selection import train_test_split, StratifiedShuffleSplit, cross_val_score
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder, StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, RandomForestClassifier, ExtraTreesClassifier
from sklearn.linear_model import SGDRegressor, BayesianRidge
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import mean_absolute_error as mae
from sklearn.metrics import accuracy_score as acc
from sklearn.metrics import make_scorer, confusion_matrix
from sklearn.dummy import DummyRegressor, DummyClassifier
```

```python
# Load the dataset
df = pd.read_csv("https://raw.githubusercontent.com/albanda/CE888/master/lab3/bank-additional-full.csv", delimiter=";")
df.info()
```

```python
# Label encode column "y"
le = preprocessing.LabelEncoder()
df["y"] = le.fit_transform(df["y"])
df.head()
```

```python
# One hot encode all categorical columns
df = pd.get_dummies(df)
df.info()
```

```python
# Drop the duration column
df.drop(['duration'], axis = 1)
```

```python
# Drop any empty rows
print(len(df))
df = df.dropna()
print(len(df))
# It seems there were no NaN values
```

```python
# Plotting a histogram of the label column "y" we can see that it is largely imbalanced with more 0s than 1s (No's then Yes')
df.hist(column="y", bins = 20)
```

```python
df["y"].value_counts()
```

```python
df.to_csv("testset.csv")
```

```python
# Create training and testing sets with a 80-20 split ensuring its stratified 
# y is the column we are trying to predict
train_set, test_set = train_test_split(df, test_size=0.2, random_state=50, stratify=df["y"])
X = train_set.loc[:, train_set.columns != 'y']
y = train_set["y"]
Y_test = test_set["y"]
X_test = test_set.loc[:, train_set.columns != 'y']
#print(len(train_set), len(test_set))
```

```python
print(sorted(sklearn.metrics.SCORERS.keys()))
```

```python
# 10 fold stratified cross-validation on training set using dummy classifier
dummy_clf = DummyClassifier(strategy='stratified')
dummy_scores = cross_val_score(dummy_clf, X, y, cv=10, scoring=make_scorer(f1_score))
print("Dummy ACC: %0.2f +/- %0.2f" % (dummy_scores.mean(), dummy_scores.std()))
```

```python
# 10 fold stratified cross-validation on training set using forest regressor with 150 trees and depth = 6
clf = RandomForestClassifier(n_estimators=150, max_depth=6)
scores = cross_val_score(clf, X, y, cv=10, scoring=make_scorer(f1_score))
print(scores)
print("F1 Score: %0.2f +/- %0.2f" % (scores.mean(), scores.std()))
```

```python
# 10 fold stratified cross-validation on training set using forest regressor with 300 trees and depth = 6
clf = RandomForestClassifier(n_estimators=300, max_depth=6)
scores = cross_val_score(clf, X, y, cv=10, scoring=make_scorer(f1_score))
print(scores)
print("F1 Score: %0.2f +/- %0.2f" % (scores.mean(), scores.std()))
```
